```python
from sklearn import datasets

# Irisの測定データの読み込み
iris = datasets.load_iris()
```

```python
print(iris)
```

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = iris.target_names[iris.target]

display(df)
```

```python
df.to_csv('/content/drive/MyDrive/share/教育資料/機械学習_いきなり実践のススメ/いきなり実践_iris/data/iris_data.csv')
```

```python
from sklearn import svm

# dfから説明変数を抽出
df_X = df[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']]
# dfから目的変数を抽出
df_y = df['target']
display(df_X.head(1), df_y.head(1))

# 機械学習に使用するモデル
model = svm.LinearSVC()
# 機械学習
model.fit(df_X, df_y)
```

```python
target_info = [[6, 2, 2, 0.3]]
predict = model.predict(target_info)
print(predict)
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# dfから説明変数を抽出
df_X = df[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']]
# dfから目的変数を抽出
df_y = df['target']
display(df_X.head(1), df_y.head(1))

X_train, X_test, y_train, y_test = train_test_split(df_X,df_y,test_size=0.1,random_state=42,stratify=df_y)

# 機械学習に使用するモデル
model = svm.LinearSVC()
# 機械学習
model.fit(X_train, y_train)

test_predict = model.predict(X_test)

print('Accuracy: %.2f' % accuracy_score(y_test, test_predict))
```
